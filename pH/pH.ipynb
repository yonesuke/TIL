```python
import pylab
import numpy as np
%matplotlib inline
```

```python
ph = np.linspace(1,13,num=10**5)
h = np.power(0.1,ph)
oh = 10**(-14)/h
x = -20*(h-oh-0.1)/(h-oh+0.1)
pylab.figure(figsize=[8,6])
pylab.xlim(0,40)
pylab.ylim(0,14)
pylab.xlabel("x")
pylab.ylabel("pH")
pylab.plot(x,ph)
pylab.savefig("ph.jpg")
```

```python
Ka = 0.1**(4.74)
ph = np.linspace(1,13,num=10**5)
h = np.power(0.1,ph)
oh = 10**(-14)/h
x = -20*(h-oh-0.1/(1+h/Ka))/(h-oh+0.1)
pylab.figure(figsize=[8,6])
pylab.xlim(0,40)
pylab.ylim(0,14)
pylab.xlabel("x")
pylab.ylabel("pH")
pylab.plot(x,ph)
pylab.savefig("ph1.jpg")
```

```python
Ka1 = 0.1**(2.15)
Ka2 = 0.1**(7.20)
Ka3 = 0.1**(12.35)
ph = np.linspace(1,13,num=10**5)
h = np.power(0.1,ph)
oh = 10**(-14)/h
hoge = 2/(1+Ka1/h+Ka1*Ka2/(h*h)+Ka1*Ka2*Ka3*(h*h*h))
hoge = hoge * Ka1 * (1/h+2*Ka2/(h*h)+3*Ka2*Ka3/(h*h*h))
x = (hoge+oh*20-h*20)/(h-oh+0.1)
pylab.figure(figsize=[8,6])
pylab.xlim(0,60)
pylab.ylim(0,14)
pylab.xlabel("x")
pylab.ylabel("pH")
pylab.plot(x,ph)
pylab.savefig("ph2.jpg")
```

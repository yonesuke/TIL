```python
from math import exp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#時間の幅
h=0.01
#ステップ数
N=10**3
#tの値
ts=[i*h for i in range(N)]
```

はじめに微分方程式の厳密解をプロットする。
$$
y(t)=\frac{1}{1+(1/y_{0}-1)e^{-t}}
$$

```python
#初期値0.5
y0=0.5
C=1/y0-1
ys_diff_05=[1/(1+C*exp(-t)) for t in ts]
#初期値1.5
y0=1.5
C=1/y0-1
ys_diff_15=[1/(1+C*exp(-t)) for t in ts]
plt.plot(ts,ys_diff_05)
plt.plot(ts,ys_diff_15)
```

次にロジスティックマップを考える。
$$
y_{n+1}=y_{n}+h(1-y_{n})y_{n}
$$

```python
plt.figure(figsize=[12,8])
#初期値0.5
y0=0.5
ys_logi_05=[y0]
for i in range(1,N):
    y=ys_logi_05[i-1]
    y_new=3.5*(1-y)*y
    ys_logi_05.append(y_new)
plt.plot(ts,ys_logi_05)
```

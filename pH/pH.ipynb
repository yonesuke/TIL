```python
import pylab
import numpy as np
%matplotlib inline
```

```python
ph = np.linspace(1,13,num=10**5)
h = np.power(0.1,ph)
oh = 10**(-14)/h
x = -20*(h-oh-0.1)/(h-oh+0.1)
```

```python
pylab.figure(figsize=[8,6])
pylab.xlim(0,40)
pylab.ylim(0,14)
pylab.xlabel("x")
pylab.ylabel("pH")
pylab.plot(x,ph)
pylab.savefig("ph.jpg")
```

```python
Ka = 0.1**(4.74)
ph = np.linspace(1,13,num=10**5)
h = np.power(0.1,ph)
oh = 10**(-14)/h
x = -20*(h-oh-0.1/(1+h/Ka))/(h-oh+0.1)
```

```python
pylab.figure(figsize=[8,6])
pylab.xlim(0,40)
pylab.ylim(0,14)
pylab.xlabel("x")
pylab.ylabel("pH")
pylab.plot(x,ph)
pylab.savefig("ph1.jpg")
```
